```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import os
```

```python
training = pd.read_csv('training_set.csv')
```

```python
train_aug_super = pd.read_csv('train_aug_new_crop_rot_gauss.csv')
```

```python
#Verigying before submit
# Create figure and axes
np.random.seed(9900)
imx = np.random.choice(os.listdir('train_new_aug_crop_rot_gauss/'))
im = cv2.imread('train_new_aug_crop_rot_gauss/' + imx)
im = cv2.cvtColor(im, cv2.COLOR_RGB2GRAY)
#im = np.expand_dims(gray, axis = -1)
fig,ax = plt.subplots(1)
x1 = train_aug_super.loc[train_aug_super['image_name'] == imx,'x1'].values[0]
x2 = train_aug_super.loc[train_aug_super['image_name'] == imx,'x2'].values[0]
y1 = train_aug_super.loc[train_aug_super['image_name'] == imx,'y1'].values[0]
y2 = train_aug_super.loc[train_aug_super['image_name'] == imx,'y2'].values[0]

# Display the image
#ax.imshow(im[0,:,:,0])
ax.imshow(im)

# Create a Rectangle patch
rect2 = patches.Rectangle((x1,y1),x2-x1,y2-y1,linewidth=1,edgecolor='r',facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect2)

plt.show()
```

```python
train_aug_super = train_aug_super[['image_name','x1','x2','y1','y2']]
```

```python
train_real_plus_train_aug_crop_rot_gauss = pd.concat([training, train_aug_super], axis = 0)
```

```python
train_real_plus_train_aug_crop_rot_gauss.to_csv('train_real_plus_train_aug_crop_rot_gauss.csv', index=False)
```
